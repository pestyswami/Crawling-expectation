"https://search.naver.com/search.naver?where=news&sm=tab_pge&query={keyword}&sort=0&photo=3&field=0&pd=3&ds={start_day}&de={end_day}&cluster_rank=1&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from{start_day}to{end_day},a:all&start={page}" url은 위와 같은 형식 <br> <br>

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib import rc
%matplotlib inline

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv("./news22_01_02.csv")

plt.figure(figsize=(15, 8))
plt.rcParams.update({'font.size': 13})

ax = df.set_index('day')['윤석열'].plot(kind='line', marker='d')
ax = df.set_index('day')['이재명'].plot(kind='line', marker='d')
ax = df.set_index('day')['안철수'].plot(kind='line', marker='d')
ax = df.set_index('day')['심상정'].plot(kind='line', marker='d')

plt.legend(loc=(0.05, 0.8))
ax.set_ylabel('기사 수')
ax.set_xlabel('날짜')
plt.show()
```

# 결론
<hr/>

## 1. 주말의 경우 기사의 양이 적어 결론을 내는데 불필요 할 것 이다.
-> 주말의 경우 평일보다 적은 기사량을 보인 것은 맞고 추가로 금요일도 적기 때문에 정확한 분석을 위해선 제외해야 한다.

<hr/>

## 2. 대선이 가까워 질 수록 정치 기사의 양이 전체적으로 증가할 것 이다.
-> 큰 차이를 보이지는 않았지만 약 10.9% 증가했다. 대선이 가까워질 수록 정치기사가 많이 나온다는 것을 수치로 확인할 수 있다.

<hr/>

## 3. 지지율이 높은 사람일수록 많은 지지자들이 궁금해하기 때문에 해당 정치인 관련된 기사들이 많이 나올것이다.
-> 정확한 수치까지는 알 수 없을지라도 기사량 그래프와 지지율 그래프가 동일한 기조를 보인 것으로 보아 여론조사전 기사의 수 만으로도 어느정도 파악할 수 있다는 것을 알 수 있다.

<hr/>

## 4. 대통령이 되고 난 후의 많은 기사량은 대통령의 지지율과 반비례할 것 이다.
-> 대선후보 지지율과 달리 국정지지율은 기사량과 계속해서 비례하지는 않는다는 것을 알 수 있다. <br> 
5월초의 많은 기사량은 5/10일날 취임했기에 취임 초기의 여파로 추측할 수 있다. <br> 
취임후 서서히 기사의 양이 감소하는 기조였는데 지지율 하락이 있기전 평소보다 많은 기사량이 나온 날이 있다는 것을 알수 있다. <br> 
6월 1주차의 지지율 하락전 6월 2일 474개의 기사가 있었고 6월 5주차의 하락은 6월 22일 351개의 기사가 있었다. <br>
이는 대통령과 관련한 이슈가 발생한날 많은 기사가 나오고 그 이슈의 영향으로 지지율이 하락한 것으로 추측할 수 있다. <br>
이슈가 있으면 당일날 기사는 당연히 많이 나오고 기조가 계속 유지되어 한동안 많은 기사가 나올줄 알았는데 사실은 당일날 많은 기사가 나온 후 서서히 기사의 양이 감소하는 것이였고 이를 파악하지 못해 잘못된 가설을 세웠다.
